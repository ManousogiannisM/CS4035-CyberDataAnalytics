```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
#Import dataset
TrainingSetdf = pd.read_csv("BATADAL_dataset03.csv", parse_dates=True, index_col='DATETIME', dayfirst=True)
EvalSetdf=pd.read_csv("BATADAL_dataset04.csv", parse_dates=True, index_col='DATETIME', dayfirst=True)
```

```python
TrainingSetdf.head(10)
y=TrainingSetdf.P_J415 #Take a Sensor
```

```python
Fs = len(y);  #sampling rate
Ts = 1.0/Fs; #sampling interval
t = np.arange(0,1,Ts) #time vector
```

```python
n = len(y)
print(n)
k = np.arange(n)
T = n/Fs
frq = k/T #two sides frequency range
```

```python
Y = np.fft.fft(y)/n #fft computing and normalization
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y),'r') #plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
plt.ylim((0,np.sort(abs(Y))[::-1][1]))
```
