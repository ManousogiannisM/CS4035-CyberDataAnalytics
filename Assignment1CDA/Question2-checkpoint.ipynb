```python
##Question2
import os
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix

cwd = os.getcwd()
dataFrame = pd.read_csv(cwd+"\data_for_student_case.csv")
```

```python
#Preprocessing
#Remove refused transactions from dataset
dataFrame = dataFrame.loc[dataFrame['simple_journal'] != 'Refused']
dataFrame.info()

#convert booking date and creation date to DataTime
dates=['bookingdate', 'creationdate']
for col in dates:
    dataFrame[col] = pd.to_datetime(dataFrame.bookingdate, format='%Y-%m-%d %H:%M:%S', errors='coerce')

#Handle the following attributes as categorical
categoricalAtrr=['issuercountrycode', 'txvariantcode', 'currencycode', 'shoppercountrycode', 'shopperinteraction',
               'cardverificationcodesupplied', 'cvcresponsecode', 'accountcode']
for col in categoricalAtrr:
    dataFrame[col] = dataFrame[col].astype('category')

def CurerncyConverter(currencyColumn):
    coversion_dict = {'SEK':0.0858,'MXN':0.038011,'AUD':0.556,'NZD':0.514,'GBP':1}
    currency = currencyColumn['currencycode']
    amount = currencyColumn['amount']
    return round(amount * coversion_dict[currency] / 100)

dataFrame['convertedAmount'] = dataFrame.apply(lambda x: CurerncyConverter(x),axis=1)
##Preprocessing stage finished
```

```python
#take a subset of the dataframe columns that are mor informative.mail id,ip id ,card id,txid,bin,creation date and booking date are removed
SimplifiedDf = dataFrame[['issuercountrycode', 'txvariantcode', 'convertedAmount', 'currencycode', 'shoppercountrycode',
                  'shopperinteraction', 'cardverificationcodesupplied', 'cvcresponsecode', 'accountcode',
                   'simple_journal']]
```

```python
#Replace ChargeBack with value 1 and settled with value 0 and convert them to integers
SimplifiedDf.loc[SimplifiedDf.simple_journal == 'Chargeback', 'simple_journal'] = 1
SimplifiedDf.loc[SimplifiedDf.simple_journal == 'Settled', 'simple_journal'] = 0
SimplifiedDf['simple_journal'] = SimplifiedDf['simple_journal'].astype('int')
```

```python
#create the labels df  extracting the values of simple journal. All values are either 0 or 1
Labels = SimplifiedDf.simple_journal
```

```python
#Remove labels to create the feature vector for the classifier
FeatureVector = SimplifiedDf.drop('simple_journal', axis=1)
FeatureVector.info()
```

```python
#Create extra columns indicating with 1 and 0 where one specific column(feature value) occurs or not
FeatureVector = pd.get_dummies(FeatureVector)
FeatureVector.info()
FeatureVector.head()
```

```python
#split our dataset in 50% training set and 50% test set randomly. random state used : 42
TrainSet, TestSet, TrainLabels, TestLabels = train_test_split(FeatureVector, Labels, test_size=0.5,
                                                              random_state=42, stratify=Labels)
```

```python
#Perform SMOTE technique for creating additional fraud data. They will result to be 20% of the non-fraud set
SyntheticSample = SMOTE(ratio=float(0.2), random_state=42)

SyntheticTrainSet, SyntheticLabels = SyntheticSample.fit_sample(TrainSet, TrainLabels)
```

```python
# Function to generate ROC curve values
def ROCvalues(classifier, trainSet, testSet, labelsTrain, labelsTest):

    # Train classifier
    classifier.fit(trainSet, labelsTrain)

    # Generate the ROC curves values
    labelProbability = classifier.predict_proba(testSet)[:, 1]
    fpr, tpr, thresholds = roc_curve(labelsTest, labelProbability)
    aucScore = roc_auc_score(labelsTest, labelProbability)

    return fpr, tpr, aucScore
```

```python
clfr = RandomForestClassifier()
clfr.fit(TrainSet, TrainLabels)
labelProbability = clfr.predict_proba(TestSet)[:, 1]
precision,recall, thresholds= precision_recall_curve(TestLabels, labelProbability)
```

```python
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2,
                 color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve-UNSMOTE')
```

```python
clfr = RandomForestClassifier()
clfr.fit(SyntheticTrainSet, SyntheticLabels)
labelProbability = clfr.predict_proba(TestSet)[:, 1]
precision,recall, thresholds= precision_recall_curve(TestLabels, labelProbability)
```

```python
plt.step(recall, precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2,
                 color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve-SMOTE')
```

```python
#plots ROC curves fro smote and unsmote version of a classifer
def PlotROCs(title, fpr, tpr, auc, fpr_smote, tpr_smote, auc_smote):

    plt.title(title)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr, color='red',
             label='AUC UNSMOTEd = %0.2f' % auc)
    plt.plot(fpr_smote, tpr_smote, color='blue',
             label='AUC SMOTEd = %0.2f' % auc_smote)

    plt.xlabel('False-Positives Rate')
    plt.ylabel('True-Positives Rate')
    plt.legend(loc="lower right")



```

```python
# Function to compare two ROC curves
def CompareROC(classifier, title, TrainSet, TestSet, SmoteSample, LabelsTrain, LabelsTest,
               labelsResample):
    fpr, tpr, auc = ROCvalues(classifier, TrainSet, TestSet, LabelsTrain, LabelsTest)
    fpr_smote, tpr_smote, auc_smote = ROCvalues(classifier, SmoteSample, TestSet, labelsResample,
                                                LabelsTest)
    PlotROCs(title, fpr, tpr, auc, fpr_smote, tpr_smote, auc_smote)
```

```python
CompareROC(LogisticRegression(), 'AUC of Logistic Classifier',
           TrainSet, TestSet, SyntheticTrainSet, TrainLabels, TestLabels, SyntheticLabels)
```

```python
CompareROC(tree.DecisionTreeClassifier(), 'AUC of Decision Tree Classifier',
           TrainSet, TestSet, SyntheticTrainSet,
           TrainLabels, TestLabels, SyntheticLabels)
```

```python
CompareROC(KNeighborsClassifier(n_neighbors=7), 'AUC of KNN Classifier',
           TrainSet, TestSet, SyntheticTrainSet, TrainLabels, TestLabels, SyntheticLabels)
```

```python
clfr = RandomForestClassifier(max_depth=4, criterion="gini", random_state=0)
clfr.fit(TrainSet, TrainLabels)
CompareROC(clfr, 'Random Forest',
           TrainSet, TestSet, SyntheticTrainSet, TrainLabels, TestLabels, SyntheticLabels)
```
