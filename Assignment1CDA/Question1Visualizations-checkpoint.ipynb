```python
cwd = os.getcwd()
dataFrame = pd.read_csv(cwd+"\data_for_student_case.csv")
print("Number of columns: {}".format(len(dataFrame.columns)))
dataFrame.info()
dataFrame.head()
dataFrame.describe()
```

```python
#Preprocessing Stage
#Remove refused transactions from dataset
dataFrame = dataFrame.loc[dataFrame['simple_journal'] != 'Refused']
dataFrame.info()
```

```python
#convert booking date and creation date to DataTime
dates=['bookingdate', 'creationdate']
for col in dates:
    dataFrame[col] = pd.to_datetime(dataFrame.bookingdate, format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
#Handle the following attributes as categorical
categoricalAtrr=['issuercountrycode', 'txvariantcode', 'currencycode', 'shoppercountrycode', 'shopperinteraction',
               'cardverificationcodesupplied', 'cvcresponsecode', 'accountcode']
for col in categoricalAtrr:
    dataFrame[col] = dataFrame[col].astype('category')
```

```python
#convert all currencies to GBP and add them to dataframe
def CurerncyConverter(currencyColumn):
    coversion_dict = {'SEK':0.0858,'MXN':0.038011,'AUD':0.556,'NZD':0.514,'GBP':1}
    currency = currencyColumn['currencycode']
    amount = currencyColumn['amount']
    return round(amount * coversion_dict[currency] / 100)
```

```python
dataFrame['convertedAmount'] = dataFrame.apply(lambda x: CurerncyConverter(x),axis=1)
##Prepocessing finished
```

```python
###QUESTION 1: VISUALIZATION
#takes three arguments as input. label(usually fraud or settled) represents the value that corresponds to the color
#variable 1 the variable in the x axis and variable2 the variable in the y axis.title is the title of the pllot
def CreateHeatmap(label,variable1,variable2, title):

    # modify data to correct format
    SplittedDataSet = dataFrame[dataFrame['simple_journal'] == label]
    DisplayData = SplittedDataSet.groupby([variable1, variable2]).size()\
        .reset_index(name='count')

    # Filter the country codes
    var2disp = list(dataFrame[variable2][dataFrame.simple_journal == 'Chargeback'].unique())
    var1disp = list(dataFrame[variable1][dataFrame.simple_journal == 'Chargeback'].unique())
    DisplayData = DisplayData[DisplayData[variable2].isin(var2disp)]
    DisplayData = DisplayData[DisplayData[variable1].isin(var1disp)]

    # create pivot and plot
    PivotData = DisplayData.pivot(index=variable1, columns=variable2, values='count')
    sns.heatmap(PivotData, cmap='plasma', linewidths=.5)
    plt.xlabel(variable2)
    plt.ylabel(variable1)
    plt.title(title)



    
```

```python
CreateHeatmap('Settled','shoppercountrycode','issuercountrycode', 'Issuer and Usage Country Codes of Settled Transactions')
```

```python

def boxplot(dataset,variablex,variabley, title):
    sns.boxplot(x=variablex, y=variabley, data=dataset)  

    plt.title(title)
    plt.xlabel('')
    plt.ylabel('Amount')
    plt.ylim(0, 450)
```
