```python
#computes the most frequent IPs that interact with the infected host
def Find10MostCommonInteractions(Dataset,infectedHost):
    mostFrequentIPs={}
    for index, row in Dataset.iterrows():
            if row["SourceIP"]==infectedHost:
                if row["DestinationIP"] in mostFrequentIPs:
                    mostFrequentIPs[row["DestinationIP"]]+=1
                else:
                    mostFrequentIPs[row["DestinationIP"]]=1
            else:
                if row["SourceIP"] in mostFrequentIPs:
                    mostFrequentIPs[row["SourceIP"]]+=1
                else:
                    mostFrequentIPs[row["SourceIP"]]=1   
    mostFrequentIPs = sorted(mostFrequentIPs.items(), key=operator.itemgetter(1))


    maxindex=len(mostFrequentIPs)
    minindex=len(mostFrequentIPs)-10
    for i in range(minindex,maxindex):
        print(mostFrequentIPs[i])                
                
    
```

```python
#add a random number between 0 and 1 to each row so that we can apply min max sampling
InfectedDataset = dataset.loc[(dataset['SourceIP']==infectedHost) | (dataset['DestinationIP']==infectedHost)]
InfectedDataset["minmax"]=0
InfectedDataset["minmax"]=InfectedDataset["minmax"].apply(lambda x:x+random.uniform(0,1))
InfectedDataset
```

```python
#apply sampling algorithm by selecting k samples from dataset and find the 10 most frequent
kList=[30,40,50,60,70,80,90,100,200,300,400,500,600,700,800]

InfectedDataset=InfectedDataset.sort_values("minmax",ascending=False)
for k in kList:
    print("K: ",k)
    
    SampleInfectedDataset=InfectedDataset.nlargest(k,'minmax')
    Find10MostCommonInteractions(SampleInfectedDataset,infectedHost)
    print(" ")
        

```

```python
################################### Count-Min Sketch################################################################
# A class that we implemented for the Count-Min Sketch algorithm
class CountMin:
    
    def __init__(self, w, d):
        self.size = w*d
        self.w = w
        self.hash_count = d
        self.cm_array =  [[0]*w for i in range(d)]
        
    def add(self, string):
        for seed in range(self.hash_count):
            result = mmh3.hash(string, seed) % self.w
            self.cm_array[seed][result] += 1
        
    def point(self, string):
        min = 9999999999
        for seed in range(self.hash_count):
            result = mmh3.hash(string, seed) % self.w
            if self.cm_array[seed][result]<min:
                min = self.cm_array[seed][result]
        return min
```

```python
#Implement Count-Min Sketch
import operator
from collections import Counter
from bitarray import bitarray
import mmh3
import math
e=math.exp(1)
```

```python
#initialize ε and δ
epsilon=0.0001
delta=0.0001
# derive w and d
w = round(e/epsilon)
d = round(math.log(1/delta))
# create count min instance
cm = CountMin(w, d)
# add to matrix
for ip in new_list:
    cm.add(ip)
#find frequency 
cm_res = {}
for ip in new_list:
    cm_res[ip] = cm.point(ip)
# Sort them according to their value to find the 10 most frequent ones
sorted_cm = sorted(cm_res.items(), key=operator.itemgetter(1), reverse = True)
sorted_cm
```
